```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Dimple\Documents\all practice_jupy\POWERBI8\netflix project\netflix_titles.csv")
print(df.head())
```

```python
#clean dataset
cleaned_df = df.copy()
```

```python
print(cleaned_df.head())
```

```python
#check cleaned dataset
print(cleaned_df.head())
print(cleaned_df.info())
```

```python
#Statistical Summary
print(cleaned_df.describe())
```

```python
#any Null Value
print(cleaned_df.isnull().sum())
```

```python
#duplicate values
print("Duplicate Rows:", cleaned_df.duplicated().sum())
```

```python
#correlation
correlation = cleaned_df.corr(numeric_only=True)
print(correlation)
```

```python
#Create a heatmap
plt.figure(figsize=(10,6))
sns.heatmap(cleaned_df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
#Histogram
cleaned_df.hist(figsize=(12,10))
plt.show()
```

```python
#Boxplot
numeric_cols = cleaned_df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=cleaned_df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
cleaned_df.to_csv("cleaned_dataset.csv", index=False)
```
